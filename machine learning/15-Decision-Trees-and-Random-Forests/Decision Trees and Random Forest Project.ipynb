```python
cat_feats=['purpose']
```

```python
final_data=pd.get_dummies(df,columns=cat_feats,drop_first=True)
final_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final_data.drop(['not.fully.paid'],axis=1), final_data['not.fully.paid'], test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred=dtree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
y_test.value_counts()
```

```python
print(confusion_matrix(y_test,pred))
print('\n\n\n')
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
error=[]
for i in range(50,250):
    rfct=RandomForestClassifier(n_estimators=i)
    rfct.fit(X_train,y_train)
    pre=rfct.predict(X_test)
    error.append(np.mean(pre != y_test))
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(50,250),error)
```

```python
rfc=RandomForestClassifier(n_estimators=160)
```

```python
rfc.fit(X_train,y_train)
```

```python
fpred=rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,fpred))
print('\n\n\n\n\n\n')
print(classification_report(y_test,fpred))
```
